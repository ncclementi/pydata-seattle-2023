```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
from matplotlib import rcParams #install in the environment
```

```python
# Add every font at the specified location - in this case Inter. 
font_dir = ["/Users/ncclementi/Documents/Inter/"]
for font in matplotlib.font_manager.findSystemFonts(font_dir):
    matplotlib.font_manager.fontManager.addfont(font)
```

```python
rcParams['font.family'] = 'Inter'
```

```python
df = pd.read_csv("memory_sampler_amm_demo_align.csv")
```

```python
df
```

```python
df["total_sec"] = pd.to_timedelta(df["0"]).dt.total_seconds()
```

```python
plt.plot(df["total_sec"], df["AMM on"]/ 2**30 , 
         '-', c="#1F5AFF", label='AMM on', lw=2.5)
plt.plot(df["total_sec"], df["AMM off"]/ 2**30 , 
         '-', c="#EF1561", label='AMM off', lw=2.5)
plt.ylabel("Cluster memory (GiB)", size=15)
plt.xlabel("Total seconds", size=15)
plt.legend(fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid()
plt.savefig("amm-ms-plot-array.svg", bbox_inches='tight');
```
